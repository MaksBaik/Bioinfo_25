Нашел 3 последовательности S белка SARS-CoV-2 в ncbi: OZ022790.1, OZ022789.1, OZ022784.1. Добавил к ним последовательность №10825 с семинара. Выравнял. Построил дерево в MEGA:

image.png

Сиквенс ковида с семинара оказался даже ближе к OZ022790.1 и OZ022784.1, чем OZ022789.1. Значит он тоже содержит ген spike белка.

Нашел 4 ортолога. На странице гена spike белка не было вкладки с ортологами, поэтому пришлось искать через blast и брать совпадения в которых не упоминается S-белок. После добавления ортологов выравнял еще раз, построил дерево:

image.png

Ортологи с MZ ушли недалеко и практически не отличаются от наших сиквенсов, скорее всего это даже не ортологи, а просто варианты белка у SARS_CoV-2. А вот последовательности с PV отличаются значительно, расположены на дереве обособленно, а значит с высокой вероятностью являются ортологами.
